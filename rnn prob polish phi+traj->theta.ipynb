```python
# Full script for Part 2: Inferring Theta given Trajectory and Phi
# MODIFIED: Includes evaluation every N batches

import pickle
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import TensorDataset, DataLoader, Subset
from matplotlib import pyplot as plt
import numpy as np
from nets.rnn import GRUNet # Assuming GRUNet is defined in nets/rnn.py
from plot_ult import initiate_plot, quickspine # Assuming these are defined
import time
import os # For checking file existence

# --- Configuration ---
datanote = 'data'
testnote = 'infer_theta_given_phi_eval_100batch' # Descriptive note
agent_name = 'ppo_baseline_0331_5cost'
data_path = f'data/{agent_name}_{datanote}'
model_save_path = f'data/{agent_name}_{datanote}_{testnote}.pt'
results_plot_path = f'results/{agent_name}_{datanote}_{testnote}_plots.png' # Save plots too

# Ensure results directory exists
os.makedirs('results', exist_ok=True)

# Hyperparameters
batch_size = 32
# Reduce num_epochs if evaluating very frequently, as total time will increase
num_epochs = 30 # ADJUST AS NEEDED - maybe fewer epochs if checking batches often
learning_rate = 0.005
hidden_size = 64
num_layers = 2
train_split_ratio = 0.8
val_split_ratio = 0.1
gradient_clip_norm = 5.0
phi_index = 3
theta_index = 9
phi_dim = 1
theta_dim = 1
output_size = 2

# Evaluation Frequency within Epochs
eval_batch_interval = 100 # Evaluate on validation set every 100 batches

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- Data Loading and Preprocessing ---
# (Identical to the previous Part 2 script - Load, Extract Phi/Theta, Normalize Theta, Augment Trajectories)
print(f"Loading data from: {data_path}")
if not os.path.exists(data_path): print(f"Error: Data file not found at {data_path}"); exit()
try:
    with open(data_path, 'rb') as f: x_data_np, ys_np = pickle.load(f)
except Exception as e: print(f"Error loading data: {e}"); exit()
print(f"Loaded {len(x_data_np)} samples.")
try:
    y_data_full = torch.stack([torch.tensor(y, dtype=torch.float32).view(-1) for y in ys_np])
    phi_data = y_data_full[:, phi_index].unsqueeze(1)
    y_data_theta = y_data_full[:, theta_index].unsqueeze(1)
except Exception as e: print(f"Error processing ys data: {e}"); exit()
theta_mean = torch.mean(y_data_theta); theta_std = torch.std(y_data_theta)
if theta_std == 0: theta_std = 1.0; print("Warning: Theta std is zero.")
y_data_theta_normalized = (y_data_theta - theta_mean) / theta_std
print(f"Target Theta normalized (mean={theta_mean:.4f}, std={theta_std:.4f}).")
try: x_data_tensors = [torch.tensor(x, dtype=torch.float32) for x in x_data_np]
except Exception as e: print(f"Error converting x_data to tensors: {e}"); exit()
padded_data = pad_sequence(x_data_tensors, batch_first=True, padding_value=0.0)
num_samples, max_seq_length, original_feature_dim = padded_data.shape
print(f"Trajectory data padded. Shape: {padded_data.shape}")
phi_data_expanded = phi_data.unsqueeze(1).expand(-1, max_seq_length, -1)
augmented_padded_data = torch.cat((padded_data, phi_data_expanded), dim=2)
augmented_feature_dim = augmented_padded_data.shape[2]
print(f"Augmented trajectory data created. Shape: {augmented_padded_data.shape}")

# --- Dataset Splitting ---
# (Identical to the previous Part 2 script)
indices = torch.randperm(num_samples)
train_cutoff = int(num_samples * train_split_ratio)
val_cutoff = train_cutoff + int(num_samples * val_split_ratio)
train_indices = indices[:train_cutoff]; val_indices = indices[train_cutoff:val_cutoff]; test_indices = indices[val_cutoff:]
print(f"Splitting data: {len(train_indices)} train, {len(val_indices)} validation, {len(test_indices)} test.")
train_data = TensorDataset(augmented_padded_data[train_indices], y_data_theta_normalized[train_indices])
val_data = TensorDataset(augmented_padded_data[val_indices], y_data_theta_normalized[val_indices])
test_data = TensorDataset(augmented_padded_data[test_indices], y_data_theta_normalized[test_indices])
num_workers = 2 if device == 'cuda' else 0
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)

# --- Model Definition ---
# (Identical to the previous Part 2 script)
model = GRUNet(input_size=augmented_feature_dim, hidden_size=hidden_size, num_layers=num_layers, output_size=output_size).to(device)
print("Model defined.")

# --- Loss Function (1D Gaussian NLL + MSE) ---
# (Identical to the previous Part 2 script)
def gaussian_1d_nll_mse(pred_params, targets):
    mu = pred_params[:, 0].unsqueeze(1)
    log_var = pred_params[:, 1].unsqueeze(1)
    log_var = torch.clamp(log_var, -10.0, 10.0)
    variance = torch.exp(log_var)
    distribution = torch.distributions.Normal(mu, (variance + 1e-8).sqrt())
    nll = -distribution.log_prob(targets)
    mean_nll = torch.mean(nll)
    mse = torch.mean((mu - targets)**2)
    total_loss = mean_nll + mse
    return total_loss

criterion = gaussian_1d_nll_mse
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# --- Helper Function for Evaluation ---
def evaluate_model(model, loader, criterion, device):
    """Runs model evaluation on the provided data loader."""
    model.eval() # Set model to evaluation mode
    total_loss = 0.0
    with torch.no_grad(): # Disable gradient calculations
        for x_batch, y_batch in loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            pred_params, _ = model(x_batch, None)
            loss = criterion(pred_params, y_batch)
            total_loss += loss.item()
    avg_loss = total_loss / len(loader)
    model.train() # Set model back to training mode
    return avg_loss

# --- Training Loop with Intermediate Evaluation ---
print(f"Starting training... Evaluating validation set every {eval_batch_interval} batches.")
train_losses_epoch = []
val_losses_epoch = [] # Store end-of-epoch validation losses
val_losses_batch = [] # Store intermediate validation losses
batch_eval_points = [] # Store (epoch, batch) for intermediate evals
best_epoch_val_loss = float('inf')
total_start_time = time.time()
global_batch_counter = 0

for epoch in range(num_epochs):
    epoch_start_time = time.time()
    model.train()
    epoch_train_loss = 0.0
    num_batches_in_epoch = len(train_loader)

    for i, (x_batch, y_batch) in enumerate(train_loader):
        global_batch_counter += 1
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)

        # Forward pass
        pred_params, _ = model(x_batch, None)

        # Calculate loss
        loss = criterion(pred_params, y_batch)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=gradient_clip_norm)
        optimizer.step()

        current_train_loss = loss.item()
        epoch_train_loss += current_train_loss

        # --- Intermediate Evaluation ---
        if global_batch_counter % eval_batch_interval == 0:
            intermediate_val_loss = evaluate_model(model, val_loader, criterion, device)
            val_losses_batch.append(intermediate_val_loss)
            batch_eval_points.append((epoch + 1, i + 1))
            print(f"  Epoch [{epoch+1}/{num_epochs}], Batch [{i+1}/{num_batches_in_epoch}], "
                  f"Train Loss (batch): {current_train_loss:.4f}, "
                  f"Val Loss (intermediate): {intermediate_val_loss:.4f}")
            # --- Final Evaluation and Visualization on Test Set ---
            # (Identical to the previous Part 2 script - Load best model based on epoch val loss, evaluate, de-normalize, plot)
            print("Loading best model for final evaluation...")
            try:
                checkpoint = torch.load(model_save_path, map_location=device)
                model.load_state_dict(checkpoint['model_state_dict'])
                theta_mean_eval = checkpoint['config'].get('theta_mean', theta_mean.item())
                theta_std_eval = checkpoint['config'].get('theta_std', theta_std.item())
                print(f"Best model from epoch {checkpoint.get('epoch', 'N/A')} loaded.")
            except FileNotFoundError:
                print(f"Error: Model file not found at {model_save_path}. Evaluating with last state.")
                theta_mean_eval = theta_mean.item()
                theta_std_eval = theta_std.item()
            except Exception as e:
                print(f"Error loading model: {e}. Evaluating with last state.")
                theta_mean_eval = theta_mean.item()
                theta_std_eval = theta_std.item()

            print("Evaluating on test set...")
            model.eval() # Ensure model is in eval mode
            all_pred_params = []; all_targets_normalized = []
            eval_start_time = time.time()
            with torch.no_grad():
                for x_batch, y_batch_norm in test_loader:
                    x_batch = x_batch.to(device)
                    pred_params, _ = model(x_batch, None)
                    all_pred_params.append(pred_params.cpu())
                    all_targets_normalized.append(y_batch_norm.cpu())
            all_pred_params = torch.cat(all_pred_params, dim=0)
            all_targets_normalized = torch.cat(all_targets_normalized, dim=0)
            pred_means_normalized = all_pred_params[:, 0]
            pred_log_vars = all_pred_params[:, 1]
            pred_std_devs_normalized = torch.sqrt(torch.exp(pred_log_vars) + 1e-8)
            pred_means_denormalized = pred_means_normalized * theta_std_eval + theta_mean_eval
            pred_std_devs_denormalized = pred_std_devs_normalized * theta_std_eval
            all_targets_denormalized = all_targets_normalized * theta_std_eval + theta_mean_eval
            eval_end_time = time.time()
            print(f"Evaluation Time: {eval_end_time - eval_start_time:.2f}s")
            mae = torch.mean(torch.abs(pred_means_denormalized - all_targets_denormalized.squeeze()))
            print(f"Test Set Mean Absolute Error (MAE) on Theta: {mae.item():.4f}")


            # --- Plotting ---
            # (Plotting code identical to previous Part 2 script - generates scatter and error bar plots)
            print("Generating plots...")
            plot_start_time = time.time()
            with initiate_plot(10, 5, 150) as fig:
                # Plot 1: Scatter True vs Pred Mean
                ax1 = fig.add_subplot(1, 2, 1)
                plot_subset_indices = np.random.choice(len(all_targets_denormalized), min(1000, len(all_targets_denormalized)), replace=False)
                ax1.scatter(all_targets_denormalized.numpy()[plot_subset_indices], pred_means_denormalized.numpy()[plot_subset_indices], alpha=0.3, s=15, label='Predictions')
                min_val = min(all_targets_denormalized.min(), pred_means_denormalized.min()); max_val = max(all_targets_denormalized.max(), pred_means_denormalized.max())
                ax1.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal (y=x)')
                ax1.set_xlabel("True $\\theta$"); ax1.set_ylabel("Predicted Mean $\\theta$"); ax1.set_title("True vs. Predicted Mean $\\theta$ (Test Set)")
                ax1.grid(True, linestyle='--', alpha=0.6); quickspine(ax1); ax1.legend()
                # Plot 2: Error Bars
                ax2 = fig.add_subplot(1, 2, 2)
                sorted_indices = torch.argsort(all_targets_denormalized.squeeze())
                sorted_targets = all_targets_denormalized.numpy().flatten()[sorted_indices]
                sorted_pred_means = pred_means_denormalized.numpy()[sorted_indices]
                sorted_pred_stds = pred_std_devs_denormalized.numpy()[sorted_indices]
                subset_size_err = min(100, len(sorted_indices)); subset_indices_err = np.round(np.linspace(0, len(sorted_indices) - 1, subset_size_err)).astype(int)
                ax2.errorbar(sorted_targets[subset_indices_err], sorted_pred_means[subset_indices_err], yerr=sorted_pred_stds[subset_indices_err], fmt='o', color='blue', ecolor='lightblue', elinewidth=1, capsize=3, markersize=4, alpha=0.7, label='Pred. Mean $\pm$ 1 Std Dev')
                ax2.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal (y=x)')
                ax2.set_xlabel("True $\\theta$"); ax2.set_ylabel("Predicted Mean $\\theta$"); ax2.set_title("Predicted Uncertainty vs. True $\\theta$")
                ax2.grid(True, linestyle='--', alpha=0.6); quickspine(ax2); ax2.legend()
                plt.tight_layout()
            plot_end_time = time.time()
            print(f"Plotting Time: {plot_end_time - plot_start_time:.2f}s")
            try: fig.savefig(results_plot_path, bbox_inches='tight'); print(f"Plots saved to: {results_plot_path}")
            except Exception as e: print(f"Error saving plots: {e}")
            plt.show()

            # Optionally plot intermediate validation losses
            if val_losses_batch:
                with initiate_plot(6, 4, 150) as fig_loss:
                    ax_loss = fig_loss.add_subplot(111)
                    # Create x-axis based on global batch count
                    batch_x_axis = [pt[0]*num_batches_in_epoch + pt[1] for pt in batch_eval_points] # Approximate global batch number
                    ax_loss.plot(batch_x_axis, val_losses_batch, marker='.', linestyle='-', label='Validation Loss (per 100 batches)')
                    ax_loss.set_xlabel("Global Batch Number (Approx)")
                    ax_loss.set_ylabel("Validation Loss")
                    ax_loss.set_title("Intermediate Validation Loss Curve")
                    ax_loss.grid(True, linestyle='--', alpha=0.6)
                    quickspine(ax_loss)
                    ax_loss.legend()
                    plt.tight_layout()
                    plt.show()

    # --- End-of-Epoch Evaluation ---
    avg_epoch_train_loss = epoch_train_loss / num_batches_in_epoch
    train_losses_epoch.append(avg_epoch_train_loss)

    # Perform standard end-of-epoch validation
    epoch_val_loss = evaluate_model(model, val_loader, criterion, device)
    val_losses_epoch.append(epoch_val_loss)

    epoch_end_time = time.time()
    print(f"Epoch [{epoch+1}/{num_epochs}] COMPLETE, "
          f"Train Loss (avg): {avg_epoch_train_loss:.4f}, Val Loss (epoch): {epoch_val_loss:.4f}, "
          f"Time: {epoch_end_time - epoch_start_time:.2f}s")

    # Save model checkpoint based on END-OF-EPOCH validation loss
    if epoch_val_loss < best_epoch_val_loss:
        best_epoch_val_loss = epoch_val_loss
        print(f"  ** New best epoch val loss: {best_epoch_val_loss:.4f}. Saving model... **")
        try:
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'best_val_loss': best_epoch_val_loss,
                'config': {
                    'augmented_feature_dim': augmented_feature_dim, 'hidden_size': hidden_size,
                    'num_layers': num_layers, 'output_size': output_size,
                    'phi_index': phi_index, 'theta_index': theta_index,
                    'gradient_clip_norm': gradient_clip_norm, 'learning_rate': learning_rate,
                    'theta_mean': theta_mean.item(), 'theta_std': theta_std.item()
                 },
            }, model_save_path)
        except Exception as e:
            print(f"Error saving model checkpoint: {e}")

total_end_time = time.time()
print(f"Total Training Time: {total_end_time - total_start_time:.2f}s")
print(f"Best End-of-Epoch Validation Loss: {best_epoch_val_loss:.4f}")





print("Script finished.")
```

```python
# =============================================================================
# Full Script: Part 2 - Infer Theta given Trajectory and Phi
# - Loads data, trains model, evaluates, generates all plots
# - MODIFIED: Full Test Set Eval + Plots run every N epochs
# =============================================================================

import pickle
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import TensorDataset, DataLoader, Subset
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm # For PDF calculation
import time
import os
# Optional: for binning if Phi is continuous
try:
    import pandas as pd
    PANDAS_AVAILABLE = True
except ImportError:
    PANDAS_AVAILABLE = False

# --- Custom Modules ---
# (Assuming GRUNet and plot_ult helpers are defined/imported as before)
try:
    from nets.rnn import GRUNet
except ImportError:
    print("Error: Cannot import GRUNet from nets.rnn."); exit() # Make essential
try:
    from plot_ult import initiate_plot, quickspine
except ImportError:
    print("Warning: plot_ult helpers not found. Using basic plotting.")
    import contextlib
    @contextlib.contextmanager
    def initiate_plot(w, h, dpi): fig = plt.figure(figsize=(w,h), dpi=dpi); yield fig
    def quickspine(ax):
        try: ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
        except AttributeError: pass
# =============================================================================

# --- Configuration ---
# File Paths and Notes
datanote = 'data'
experiment_note = 'infer_theta_given_phi_eval_500epoch' # Descriptive note
agent_name = 'ppo_baseline_0331_5cost'
data_path = f'data/{agent_name}_{datanote}'
model_save_path = f'data/{agent_name}_{datanote}_{experiment_note}.pt' # Best model based on val loss
results_dir = 'results'
os.makedirs(results_dir, exist_ok=True)

# --- !! Key Control Flag !! ---
DO_TRAINING = True # Set to False to load 'model_save_path' and only run final evaluation

# Model Hyperparameters
hidden_size = 64
num_layers = 2

# Training Hyperparameters
num_epochs = 15 # << Set to desired total epochs >>
batch_size = 32
learning_rate = 0.005
gradient_clip_norm = 5.0

# Data Parameters
phi_index = 3; theta_index = 9; phi_dim = 1; theta_dim = 1; output_size = 2

# Dataset Splitting
train_split_ratio = 0.8; val_split_ratio = 0.1

# --- !! Evaluation Frequency during Training !! ---
full_eval_epoch_interval = 500 # Run full test eval & plots every N epochs

# Plotting Configuration
max_phis_to_plot_dist = 12; num_dist_plot_cols = 4

# --- Device Configuration ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
num_workers = 2 if device == torch.device('cuda') else 0

# --- Data Loading and Preprocessing ---
print(f"\n--- Loading and Preprocessing Data ---")
# ... (Data loading, extraction, normalization, augmentation code - same as before) ...
print(f"Loading data from: {data_path}")
if not os.path.exists(data_path): print(f"Error: Data file not found at {data_path}"); exit()
try:
    with open(data_path, 'rb') as f: x_data_np, ys_np = pickle.load(f)
except Exception as e: print(f"Error loading data: {e}"); exit()
print(f"Loaded {len(x_data_np)} samples.")
try:
    y_data_full = torch.stack([torch.tensor(y, dtype=torch.float32).view(-1) for y in ys_np])
    phi_data = y_data_full[:, phi_index].unsqueeze(1)
    y_data_theta = y_data_full[:, theta_index].unsqueeze(1)
except Exception as e: print(f"Error processing ys data: {e}"); exit()
theta_mean = torch.mean(y_data_theta); theta_std = torch.std(y_data_theta)
if theta_std.item() == 0: theta_std = torch.tensor(1.0); print("Warning: Theta std is zero.")
y_data_theta_normalized = (y_data_theta - theta_mean) / theta_std
print(f"Target Theta normalized (mean={theta_mean.item():.4f}, std={theta_std.item():.4f}).")
try: x_data_tensors = [torch.tensor(x, dtype=torch.float32) for x in x_data_np]
except Exception as e: print(f"Error converting x_data to tensors: {e}"); exit()
padded_data = pad_sequence(x_data_tensors, batch_first=True, padding_value=0.0)
num_samples, max_seq_length, original_feature_dim = padded_data.shape
print(f"Trajectory data padded. Shape: {padded_data.shape}")
phi_data_expanded = phi_data.unsqueeze(1).expand(-1, max_seq_length, -1)
augmented_padded_data = torch.cat((padded_data, phi_data_expanded), dim=2)
augmented_feature_dim = augmented_padded_data.shape[2]
print(f"Augmented trajectory data created. Shape: {augmented_padded_data.shape}")


# --- Dataset Splitting & DataLoaders ---
print("\n--- Splitting Data and Creating DataLoaders ---")
# ... (Splitting and DataLoader creation code - same as before) ...
indices = torch.randperm(num_samples)
train_cutoff = int(num_samples * train_split_ratio); val_cutoff = train_cutoff + int(num_samples * val_split_ratio)
train_indices = indices[:train_cutoff]; val_indices = indices[train_cutoff:val_cutoff]; test_indices = indices[val_cutoff:]
print(f"Split: {len(train_indices)} train, {len(val_indices)} validation, {len(test_indices)} test.")
train_data = TensorDataset(augmented_padded_data[train_indices], y_data_theta_normalized[train_indices])
val_data = TensorDataset(augmented_padded_data[val_indices], y_data_theta_normalized[val_indices])
test_data = TensorDataset(augmented_padded_data[test_indices], y_data_theta_normalized[test_indices])
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
print("DataLoaders created.")


# --- Model Definition ---
print("\n--- Defining Model ---")
model = GRUNet(input_size=augmented_feature_dim, hidden_size=hidden_size, num_layers=num_layers, output_size=output_size).to(device)
print(f"Model: {model.__class__.__name__} defined and moved to {device}.")

# --- Loss Function & Optimizer ---
# (Same as before)
def gaussian_1d_nll_mse(pred_params, targets):
    mu = pred_params[:, 0].unsqueeze(1); log_var = pred_params[:, 1].unsqueeze(1)
    log_var = torch.clamp(log_var, -10.0, 10.0); variance = torch.exp(log_var)
    distribution = torch.distributions.Normal(mu, (variance + 1e-8).sqrt())
    nll = -distribution.log_prob(targets); mean_nll = torch.mean(nll)
    mse = torch.mean((mu - targets)**2); total_loss = mean_nll + mse
    return total_loss
criterion = gaussian_1d_nll_mse
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
print("Loss function and optimizer defined.")

# --- Helper Function for Validation Loss (Quick Check) ---
def get_validation_loss(model, loader, criterion, device):
    model.eval(); total_loss = 0.0
    with torch.no_grad():
        for x_batch, y_batch in loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            pred_params, _ = model(x_batch, None); loss = criterion(pred_params, y_batch)
            total_loss += loss.item()
    avg_loss = total_loss / max(1, len(loader))
    model.train(); return avg_loss # Return to train mode


# =============================================================================
# --- Evaluation and Plotting Function (Called Periodically) ---
# =============================================================================
def run_full_evaluation_and_plotting(model_to_eval, epoch_num, is_final_eval=False):
    """
    Runs full evaluation on the test set and generates all plots.
    Uses the provided model_to_eval state.
    """
    print(f"\n--- Running Full Evaluation & Plotting @ Epoch {epoch_num} {'(Final Best Model)' if is_final_eval else ''} ---")
    model_to_eval.eval() # Ensure eval mode

    # --- Run Inference on Test Set ---
    all_pred_params = []; all_targets_normalized = []
    eval_start_time = time.time()
    with torch.no_grad():
        for x_batch, y_batch_norm in test_loader:
            x_batch = x_batch.to(device)
            pred_params, _ = model_to_eval(x_batch, None) # Use the passed model
            all_pred_params.append(pred_params.cpu())
            all_targets_normalized.append(y_batch_norm.cpu())

    if not all_pred_params:
        print("Error: No predictions generated during full evaluation.")
        return # Exit this function if no predictions

    all_pred_params = torch.cat(all_pred_params, dim=0)
    all_targets_normalized = torch.cat(all_targets_normalized, dim=0)

    # --- Calculate and De-normalize prerequisite variables ---
    # Use global theta_mean/std for de-normalization if checkpoint loading failed earlier for final eval
    theta_mean_eval = config.get('theta_mean', theta_mean.item()) if 'config' in locals() else theta_mean.item()
    theta_std_eval = config.get('theta_std', theta_std.item()) if 'config' in locals() else theta_std.item()

    pred_means_normalized = all_pred_params[:, 0].numpy()
    pred_log_vars = all_pred_params[:, 1].numpy()
    pred_std_devs_normalized = np.sqrt(np.exp(pred_log_vars) + 1e-8)
    pred_means_denormalized = pred_means_normalized * theta_std_eval + theta_mean_eval
    pred_std_devs_denormalized = pred_std_devs_normalized * theta_std_eval
    all_targets_denormalized = (all_targets_normalized.numpy() * theta_std_eval + theta_mean_eval)
    if all_targets_denormalized.ndim > 1 and all_targets_denormalized.shape[1] == 1:
        all_targets_denormalized = all_targets_denormalized.squeeze() # Ensure 1D target array

    test_phi_values = None
    if 'phi_data' in globals() and 'test_indices' in globals():
         test_phi_values = phi_data[test_indices].squeeze().numpy()
    else: print("Warning: phi_data or test_indices missing for full eval.")

    errors = None
    if pred_means_denormalized is not None and all_targets_denormalized is not None:
         try: errors = np.abs(pred_means_denormalized - all_targets_denormalized)
         except ValueError as e: print(f"Error calculating errors in full eval: {e}")

    eval_end_time = time.time()
    print(f"Evaluation Time: {eval_end_time - eval_start_time:.2f}s")

    if errors is not None:
        mae = np.mean(errors)
        print(f"Test Set MAE: {mae:.4f}")
    else: print("MAE calculation skipped.")

    # --- Plotting Section ---
    plotting_prerequisites_ok = all(v is not None for v in [all_targets_denormalized, pred_means_denormalized, pred_std_devs_denormalized, test_phi_values, errors])

    if not plotting_prerequisites_ok:
        print("Skipping plotting due to missing prerequisite variables during full evaluation.")
        return # Exit function

    print("Generating plots...")
    plot_overall_start_time = time.time()
    eval_suffix = f"epoch_{epoch_num}" if not is_final_eval else "final"

    # --- Plotting Section 1 & 2: Overall Performance and Uncertainty ---
    # (Plotting code - same as before)
    # ... (Copy the plotting code for Overall Perf/Uncertainty here) ...
    # (Modify save path to include epoch/final suffix)
    min_val = min(all_targets_denormalized.min(), pred_means_denormalized.min()); max_val = max(all_targets_denormalized.max(), pred_means_denormalized.max())
    val_range = max_val - min_val; plot_min = min_val - 0.1 * val_range; plot_max = max_val + 0.1 * val_range
    overall_plot_path = os.path.join(results_dir, f"{agent_name}_{datanote}_{experiment_note}_overall_perf_{eval_suffix}.png")
    fig_overall, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), dpi=150)
    plot_subset_indices = np.random.choice(len(all_targets_denormalized), min(1000, len(all_targets_denormalized)), replace=False)
    ax1.scatter(all_targets_denormalized[plot_subset_indices], pred_means_denormalized[plot_subset_indices], alpha=0.3, s=15, label='Predictions')
    ax1.plot([plot_min, plot_max], [plot_min, plot_max], 'r--', label='Ideal (y=x)'); ax1.set_xlabel("True $\\theta$"); ax1.set_ylabel("Predicted Mean $\\theta$"); ax1.set_title(f"Overall Perf. @ Epoch {epoch_num}")
    ax1.grid(True, linestyle='--', alpha=0.6); quickspine(ax1); ax1.legend(); ax1.set_xlim(plot_min, plot_max); ax1.set_ylim(plot_min, plot_max)
    sorted_indices = np.argsort(all_targets_denormalized.squeeze()); sorted_targets = all_targets_denormalized.squeeze()[sorted_indices]; sorted_pred_means = pred_means_denormalized.squeeze()[sorted_indices]; sorted_pred_stds = pred_std_devs_denormalized.squeeze()[sorted_indices]
    subset_size_err = min(100, len(sorted_indices)); subset_indices_err = np.round(np.linspace(0, len(sorted_indices) - 1, subset_size_err)).astype(int)
    ax2.errorbar(sorted_targets[subset_indices_err], sorted_pred_means[subset_indices_err], yerr=sorted_pred_stds[subset_indices_err], fmt='o', color='blue', ecolor='lightblue', elinewidth=1, capsize=3, markersize=4, alpha=0.7, label='Pred. Mean $\pm$ 1 Std Dev')
    ax2.plot([plot_min, plot_max], [plot_min, plot_max], 'r--', label='Ideal (y=x)'); ax2.set_xlabel("True $\\theta$"); ax2.set_ylabel("Predicted Mean $\\theta$"); ax2.set_title(f"Overall Uncertainty @ Epoch {epoch_num}")
    ax2.grid(True, linestyle='--', alpha=0.6); quickspine(ax2); ax2.legend(); ax2.set_xlim(plot_min, plot_max); ax2.set_ylim(plot_min, plot_max)
    plt.tight_layout(); plt.close(fig_overall) # Close to avoid displaying now
    try: fig_overall.savefig(overall_plot_path, bbox_inches='tight'); print(f"Overall plots saved: {overall_plot_path}")
    except Exception as e: print(f"Error saving overall plots: {e}")

    # --- Plotting Section 3: Performance Analysis vs. Phi ---
    # (Combined Analysis + Plotting code - same as before)
    # ... (Copy the Analysis + Plotting code for Perf vs Phi here) ...
    # (Modify save path to include epoch/final suffix)
    perf_vs_phi_plot_path = os.path.join(results_dir, f"{agent_name}_{datanote}_{experiment_note}_avg_pred_vs_phi_{eval_suffix}.png")
    unique_phis = np.unique(test_phi_values); analysis_type = 'unique'; num_unique_phis_threshold = 20; plot_phis = None; plot_avg_pred_mean = None; plot_avg_std = None; plot_counts = None
    if PANDAS_AVAILABLE and len(unique_phis) > num_unique_phis_threshold: # Binning Path
        num_bins = 10
        try: df_results = pd.DataFrame({'phi': test_phi_values, 'pred_mean': pred_means_denormalized, 'pred_std': pred_std_devs_denormalized}); df_results['phi_bin'] = pd.cut(df_results['phi'], bins=num_bins); grouped_perf = df_results.groupby('phi_bin', observed=False).agg(avg_pred_mean_theta=('pred_mean', 'mean'), mean_pred_std=('pred_std', 'mean'), count=('phi', 'size')).reset_index(); grouped_perf['phi_center'] = grouped_perf['phi_bin'].apply(lambda x: x.mid); plot_phis = grouped_perf['phi_center'].values; plot_avg_pred_mean = grouped_perf['avg_pred_mean_theta'].values; plot_avg_std = grouped_perf['mean_pred_std'].values; plot_counts = grouped_perf['count'].values; analysis_type = 'binned'
        except Exception as e: print(f"Pandas binning failed: {e}. Fallback."); plot_phis = unique_phis; analysis_type = 'unique'
    if analysis_type == 'unique': # Unique Value Path
        if plot_phis is None: plot_phis = unique_phis
        if len(plot_phis) > 100: print("Limiting unique phi analysis to 100 values."); indices = np.round(np.linspace(0, len(plot_phis)-1, 100)).astype(int); plot_phis = plot_phis[indices]
        per_phi_avg_pred_mean_list, per_phi_avg_std_list, per_phi_counts_list = [], [], []
        for p in plot_phis:
            indices = np.where(test_phi_values == p)[0]; count = len(indices)
            if count > 0: phi_avg_pred_mean = np.mean(pred_means_denormalized[indices]); phi_avg_std = np.mean(pred_std_devs_denormalized[indices]); per_phi_avg_pred_mean_list.append(phi_avg_pred_mean); per_phi_avg_std_list.append(phi_avg_std); per_phi_counts_list.append(count)
            else: per_phi_avg_pred_mean_list.append(np.nan); per_phi_avg_std_list.append(np.nan); per_phi_counts_list.append(0)
        plot_avg_pred_mean = np.array(per_phi_avg_pred_mean_list); plot_avg_std = np.array(per_phi_avg_std_list); plot_counts = np.array(per_phi_counts_list)
    if plot_phis is not None: # Plotting
        fig_perf_phi, (ax_left, ax_std) = plt.subplots(1, 2, figsize=(12, 5), dpi=150); sizes = 10 + np.sqrt(np.maximum(plot_counts, 1)) * 2
        scatter_left = ax_left.scatter(plot_phis, plot_avg_pred_mean, s=sizes, alpha=0.7, label=f'Avg Pred $\\mu_\\theta$ ({analysis_type} Phi)', cmap='viridis', c=plot_counts); cbar_left = fig_perf_phi.colorbar(scatter_left, ax=ax_left); cbar_left.set_label('Sample Count'); ax_left.set_xlabel("True $\\phi$"); ax_left.set_ylabel("Avg Pred Mean $\\theta$"); ax_left.set_title(f"Avg Pred $\\theta$ vs $\\phi$ @ Ep {epoch_num}"); ax_left.grid(True, linestyle='--', alpha=0.6); quickspine(ax_left); ax_left.legend()
        scatter_std = ax_std.scatter(plot_phis, plot_avg_std, s=sizes, alpha=0.7, label=f'Avg Pred Std ({analysis_type} Phi)', cmap='viridis', c=plot_counts); cbar_std = fig_perf_phi.colorbar(scatter_std, ax=ax_std); cbar_std.set_label('Sample Count'); ax_std.set_xlabel("True $\\phi$"); ax_std.set_ylabel("Avg Pred Std Dev $\\theta$"); ax_std.set_title(f"Model Uncertainty vs $\\phi$ @ Ep {epoch_num}"); ax_std.grid(True, linestyle='--', alpha=0.6); quickspine(ax_std); ax_std.legend()
        plt.suptitle(f"Performance vs. Phi @ Epoch {epoch_num}"); plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.close(fig_perf_phi)
        try: fig_perf_phi.savefig(perf_vs_phi_plot_path, bbox_inches='tight'); print(f"Perf vs Phi plots saved: {perf_vs_phi_plot_path}")
        except Exception as e: print(f"Error saving perf vs phi plots: {e}")

    # --- Plotting Section 4: Representative Prediction Distributions per Phi ---
    # (Plotting code - same as before)
    # ... (Copy the Plotting code for Distributions per Phi here) ...
    # (Modify save path to include epoch/final suffix)
    dist_phi_plot_path = os.path.join(results_dir, f"{agent_name}_{datanote}_{experiment_note}_pred_dist_per_phi_{eval_suffix}.png")
    unique_phis_dist = np.unique(test_phi_values); print(f"Found {len(unique_phis_dist)} unique Phi values for distribution plots.")
    if len(unique_phis_dist) > max_phis_to_plot_dist: print(f"Selecting {max_phis_to_plot_dist} representative Phi values to plot distributions."); phi_indices_to_plot = np.round(np.linspace(0, len(unique_phis_dist) - 1, max_phis_to_plot_dist)).astype(int); selected_phis_dist = unique_phis_dist[phi_indices_to_plot]
    else: print(f"Plotting distributions for all {len(unique_phis_dist)} unique Phi values."); selected_phis_dist = unique_phis_dist
    num_plots = len(selected_phis_dist)
    if num_plots > 0:
        num_rows_dist_phi = (num_plots + num_dist_plot_cols - 1) // num_dist_plot_cols; fig_dist_phi, axes = plt.subplots(num_rows_dist_phi, num_dist_plot_cols, figsize=(num_dist_plot_cols * 4.5, num_rows_dist_phi * 3.5), squeeze=False); axes = axes.flatten(); plot_counter = 0
        for phi_val in selected_phis_dist:
            ax = axes[plot_counter]; indices_for_phi = np.where(test_phi_values == phi_val)[0]
            if len(indices_for_phi) > 0:
                errors_for_phi = errors[indices_for_phi]; median_error_local_idx = np.argsort(errors_for_phi)[len(errors_for_phi) // 2]; representative_idx = indices_for_phi[median_error_local_idx]
                true_theta = all_targets_denormalized[representative_idx]; pred_mean = pred_means_denormalized[representative_idx]; pred_std = max(pred_std_devs_denormalized[representative_idx], 1e-4); sample_error = errors[representative_idx]
                plot_std_range = 3.5; theta_min = pred_mean - plot_std_range * pred_std; theta_max = pred_mean + plot_std_range * pred_std; plot_min_dist = min(theta_min, true_theta) - 0.5 * pred_std; plot_max_dist = max(theta_max, true_theta) + 0.5 * pred_std; theta_values = np.linspace(plot_min_dist, plot_max_dist, 200)
                pdf_values = norm.pdf(theta_values, loc=pred_mean, scale=pred_std)
                ax.plot(theta_values, pdf_values, color='black', linestyle='-', label=f'Pred. Dist.'); ax.fill_between(theta_values, pdf_values, alpha=0.2, color='lightblue'); ax.axvline(pred_mean, color='blue', linestyle='--', linewidth=2, label=f'Pred Mean ({pred_mean:.2f})'); ax.axvline(true_theta, color='red', linestyle='-', linewidth=2, label=f'True Value ({true_theta:.2f})')
                ax.text(0.95, 0.95, f'$\\phi={phi_val:.3f}$\nError={sample_error:.2f}', transform=ax.transAxes, ha='right', va='top', fontsize=9, bbox=dict(boxstyle='round,pad=0.3', fc='wheat', alpha=0.5)); ax.set_xlabel("Parameter $\\theta$"); ax.set_ylabel("Probability Density"); ax.set_title(f"Rep Sample $\\phi={phi_val:.3f}$ @ Ep {epoch_num}"); ax.grid(True, linestyle='--', alpha=0.4); ax.legend(fontsize=8, loc='upper left'); quickspine(ax); ax.set_ylim(bottom=0); ax.set_yticks([])
                plot_counter += 1
            else: ax.set_title(f"No samples for $\\phi={phi_val:.3f}$"); ax.text(0.5, 0.5, "No Data", ha='center', va='center')
        for j in range(plot_counter, len(axes)): fig_dist_phi.delaxes(axes[j]) # Hide unused
        plt.suptitle(f"Representative Predicted Distributions for Theta @ Epoch {epoch_num}"); plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.close(fig_dist_phi)
        try: fig_dist_phi.savefig(dist_phi_plot_path, bbox_inches='tight'); print(f"Distribution plots saved: {dist_phi_plot_path}")
        except Exception as e: print(f"Error saving distribution plots: {e}")

    plot_end_time = time.time()
    print(f"Total Plotting Time for Epoch {epoch_num}: {plot_end_time - eval_start_time:.2f}s (includes eval)") # Time includes eval now
    model_to_eval.train() # Return model to training mode if it was passed
# =============================================================================


# --- Training Loop ---
if DO_TRAINING:
    print("\n--- Starting Training ---")
    train_losses_epoch = []
    val_losses_epoch = []
    best_epoch_val_loss = float('inf')
    total_start_time = time.time()

    for epoch in range(num_epochs):
        epoch_start_time = time.time()
        model.train() # Ensure model is in training mode
        epoch_train_loss = 0.0
        num_batches_in_epoch = len(train_loader)

        for i, (x_batch, y_batch) in enumerate(train_loader):
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            # Forward, Loss, Backward, Optimize steps...
            pred_params, _ = model(x_batch, None); loss = criterion(pred_params, y_batch)
            optimizer.zero_grad(); loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=gradient_clip_norm)
            optimizer.step()
            epoch_train_loss += loss.item()
            if i%500==0:
                run_full_evaluation_and_plotting(model, epoch_num=epoch+1, is_final_eval=False)

        # --- End-of-Epoch Validation (for Model Saving) ---
        avg_epoch_train_loss = epoch_train_loss / num_batches_in_epoch
        train_losses_epoch.append(avg_epoch_train_loss)
        epoch_val_loss = get_validation_loss(model, val_loader, criterion, device) # Quick val check
        val_losses_epoch.append(epoch_val_loss)
        epoch_end_time = time.time()
        print(f"Epoch [{epoch+1}/{num_epochs}] | Train Loss: {avg_epoch_train_loss:.4f} | Val Loss: {epoch_val_loss:.4f} | Time: {epoch_end_time - epoch_start_time:.2f}s")

        # Save model checkpoint based on validation loss
        if epoch_val_loss < best_epoch_val_loss:
            best_epoch_val_loss = epoch_val_loss
            print(f"  ** New best epoch val loss: {best_epoch_val_loss:.4f}. Saving model to {model_save_path} **")
            # ... (model saving dictionary logic - same as before) ...
            try: torch.save({'epoch': epoch + 1,'model_state_dict': model.state_dict(), 'optimizer_state_dict': optimizer.state_dict(), 'best_val_loss': best_epoch_val_loss, 'config': {'augmented_feature_dim': augmented_feature_dim, 'hidden_size': hidden_size, 'num_layers': num_layers, 'output_size': output_size, 'phi_index': phi_index, 'theta_index': theta_index, 'gradient_clip_norm': gradient_clip_norm, 'learning_rate': learning_rate, 'theta_mean': theta_mean.item(), 'theta_std': theta_std.item()}}, model_save_path)
            except Exception as e: print(f"Error saving model: {e}")




    total_end_time = time.time()
    print(f"\n--- Training Finished ---")
    print(f"Total Training Time: {total_end_time - total_start_time:.2f}s")
    print(f"Best End-of-Epoch Validation Loss recorded: {best_epoch_val_loss:.4f}")
else:
    print("\n--- Skipping Training ---")
    if not os.path.exists(model_save_path):
        print(f"Error: Model file not found at {model_save_path} and training was skipped.")
        exit()

# =============================================================================
# --- Final Evaluation using BEST Saved Model ---
# =============================================================================
print("\n--- Final Evaluation using Best Saved Model (based on validation loss) ---")
if not os.path.exists(model_save_path):
     print(f"Error: Model file not found at {model_save_path}. Cannot run final evaluation.")
else:
    print(f"Loading best model from: {model_save_path}")
    try:
        checkpoint = torch.load(model_save_path, map_location=device)
        # Load config first to get normalization constants
        config = checkpoint.get('config', {})
        theta_mean_eval = config.get('theta_mean', theta_mean.item()) # Use loaded or current mean/std
        theta_std_eval = config.get('theta_std', theta_std.item())
        # Initialize a fresh model instance with correct architecture before loading state_dict
        final_model = GRUNet(input_size=augmented_feature_dim, hidden_size=hidden_size, num_layers=num_layers, output_size=output_size).to(device)
        final_model.load_state_dict(checkpoint['model_state_dict'])
        loaded_epoch = checkpoint.get('epoch', 'N/A')
        print(f"Best model from epoch {loaded_epoch} loaded successfully.")
        # Run final evaluation and plotting using the loaded best model
        run_full_evaluation_and_plotting(final_model, epoch_num=loaded_epoch, is_final_eval=True)
    except Exception as e:
        print(f"Error loading best model or running final evaluation: {e}")
        print("Please check model save path and file integrity.")


print("\nScript finished.")
```
